```python
import pandas as pd
import numpy as np
```

```python
plantas=pd.read_csv('/content/plant_growth_data.csv')
plantas
```

```python
plantas.describe()
```

```python
plantas["Soil_Type"].value_counts()
```

```python
plantas["Soil_Type"]=plantas["Soil_Type"].replace("clay",1)
plantas["Soil_Type"]=plantas["Soil_Type"].replace("loam",2)
plantas["Soil_Type"]=plantas["Soil_Type"].replace("sandy",3)
plantas
```

```python
plantas
```

```python
plantas["Water_Frequency"].value_counts()
```

```python
plantas["Water_Frequency"]=plantas["Water_Frequency"].replace("daily",1)
plantas["Water_Frequency"]=plantas["Water_Frequency"].replace("bi-weekly",2)
plantas["Water_Frequency"]=plantas["Water_Frequency"].replace("weekly",3)
plantas
```

```python
plantas["Fertilizer_Type"].value_counts()
```

```python
plantas["Fertilizer_Type"]=plantas["Fertilizer_Type"].replace("none",1)
plantas["Fertilizer_Type"]=plantas["Fertilizer_Type"].replace("chemical",2)
plantas["Fertilizer_Type"]=plantas["Fertilizer_Type"].replace("organic",3)
plantas
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.pairplot(plantas,hue="Growth_Milestone")
```

```python
X=plantas[["Water_Frequency",	"Fertilizer_Type","Temperature",	"Humidity"]]
y=plantas["Growth_Milestone"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
svC=SVC(kernel="linear",random_state=0)
#svC=SVC(kernel="poly",random_state=0)
#svC=SVC(kernel="rbf",random_state=0)
#svC=SVC(kernel="sigmoid",random_state=0)



```

```python
svC.fit(X_train,y_train)
```

```python
y_predTest=svC.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
mx=confusion_matrix(y_test,y_predTest)
mx
```

```python
sns.heatmap(mx,annot=True)
```

```python
print(classification_report(y_test,y_predTest))
```
