```python
import pandas as pd
```

```python
plantas=pd.read_csv('/content/plant_growth_data.csv')
```

```python
plantas
```

```python
cambiarSuelo=({'clay':0,'sandy':1,'loam':2})
plantas['Soil_Type']=plantas['Soil_Type'].map(cambiarSuelo)
```

```python
cambiarAgua=({'daily':0,'bi-weekly':1,'weekly':2})
plantas['Water_Frequency']=plantas['Water_Frequency'].map(cambiarAgua)
```

```python
cambiarF=({'none':0,'chemical':1,'organic':2})
plantas['Fertilizer_Type']=plantas['Fertilizer_Type'].map(cambiarF)
```

```python
plantas
```

```python
#!pip install lightgbm
```

```python
import lightgbm as lbg
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=plantas.drop('Growth_Milestone',axis=1)
y=plantas['Growth_Milestone']
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
y_test
```

```python
modeloLgbm=lbg.LGBMClassifier(random_state=42)
```

```python
modeloLgbm.fit(x_train,y_train)
```

```python
ypredtest=modeloLgbm.predict(x_test)
```

```python
ypredtest
```

```python
from sklearn.metrics import mean_squared_error,confusion_matrix,accuracy_score,classification_report
```

```python
mx=confusion_matrix(y_test,ypredtest)
mx
```

```python
print(classification_report(y_test,ypredtest))
```
