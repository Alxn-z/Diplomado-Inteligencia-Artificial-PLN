```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
vidrio=pd.read_csv('/content/glass.csv')
vidrio.head()
```

```python
sns.pairplot(vidrio,hue='Type')
```

```python
X=X1=vidrio.drop('Type',axis=1)
y=vidrio['Type']
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(X)
```

```python
X_scaled=scaler.transform(X)
X_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
#pca = PCA(n_components=2)
pca = PCA(n_components=3)
```

```python
pca.fit(X_scaled)
```

```python
print(pca.explained_variance_ratio_)
```

```python
print(pca.singular_values_)
```

```python
#print(pd.DataFrame(pca.components_,columns=X.columns))
print(pd.DataFrame(pca.components_,columns=X1.columns))
```

```python
x_scaledPandas=pd.DataFrame(X_scaled,columns=X.columns)
x_scaledPandas
```

```python
#de acuerdo al PCA, estos son los atributos que más contienen información
X=x_scaledPandas[["RI","Al","Ca","Mg"]]
y=vidrio['Type']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=101)
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
#falta configurar el perceptrón
clf = MLPClassifier(random_state=1,hidden_layer_sizes=(100,100,100),  max_iter=300).fit(x_train, y_train)
```

```python
ypre=clf.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
mx=confusion_matrix(y_test,ypre)
mx
```

```python
sns.heatmap(mx,annot=True)
```

```python
print(classification_report(y_test,ypre))
```

```python
from sklearn.datasets import load_iris

iris = load_iris(as_frame=True)
print(iris.keys())
```

```python
import seaborn as sns

# Rename classes using the iris target names
iris.frame["target"] = iris.target_names[iris.target]
_ = sns.pairplot(iris.frame, hue="target")
```

```python
X_reduced = PCA(n_components=3).fit_transform(iris.data)
```

```python
X_reduced
```

```python
import matplotlib.pyplot as plt

# unused but required import for doing 3d projections with matplotlib < 3.2
fig = plt.figure(1, figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn.decomposition import PCA
scatter = ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=iris.target,
    s=40,
)
ax.set(
    title="First three PCA dimensions",
    xlabel="1st Eigenvector",
    ylabel="2nd Eigenvector",
    zlabel="3rd Eigenvector",
)
ax.xaxis.set_ticklabels([])
ax.yaxis.set_ticklabels([])
ax.zaxis.set_ticklabels([])

# Add a legend
legend1 = ax.legend(
    scatter.legend_elements()[0],
    iris.target_names.tolist(),
    loc="upper right",
    title="Classes",
)
ax.add_artist(legend1)

plt.show()
```
