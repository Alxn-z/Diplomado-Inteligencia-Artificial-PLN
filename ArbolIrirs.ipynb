```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn import metrics, model_selection
```

```python
iris= load_iris()
```

```python
iris
```

```python
x,y= iris.data, iris.target
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
#máximo número de niveles del árbol, hiperparámetro
#arbol= tree.DecisionTreeClassifier(max_leaf_nodes=5)
arbol= tree.DecisionTreeClassifier()
```

```python
#hace la matemática,con el algoritmo"ID3", ganancia de información
ArbolEntrenado= arbol.fit(x_train, y_train)
```

```python
import seaborn as sns
import pandas as pd
```

```python
datFrameiris= pd.DataFrame(iris.data, columns=iris.feature_names)
datFrameiris
```

```python
sns.pairplot(datFrameiris)
```

```python
tree.plot_tree(ArbolEntrenado)
```

```python
#predicción
ypredTesta= ArbolEntrenado.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
mx= confusion_matrix(y_test, ypredTesta)
mx
```

```python
sns.heatmap(mx, annot=True)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, ypredTesta))
```

```python
#máximo número de niveles del árbol, hiperparámetro
#arbol= tree.DecisionTreeClassifier(max_leaf_nodes=4, criterion='entropy')
arbol= tree.DecisionTreeClassifier(criterion='entropy')
```

```python
ArbolEntrenadoEntropi= arbol.fit(x_train, y_train)
```

```python
tree.plot_tree(ArbolEntrenadoEntropi)
```

```python
ypredTestaEntropi= ArbolEntrenadoEntropi.predict(x_test)
```

```python
mxEntropi= confusion_matrix(y_test, ypredTestaEntropi)
mxEntropi
```

```python
print(classification_report(y_test, ypredTestaEntropi))
```
