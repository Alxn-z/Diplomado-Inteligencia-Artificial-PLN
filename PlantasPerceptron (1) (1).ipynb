```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(plantas.drop('Growth_Milestone', axis=1), plantas['Growth_Milestone'], test_size=0.2, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
mlp=MLPClassifier(hidden_layer_sizes=(100,200,100),solver='adam',max_iter=500)
```

```python
mlp.fit(X_train,y_train)
```

```python
ypredTest=mlp.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
mx=confusion_matrix(y_test,ypredTest)
mx
```

```python
print(classification_report(y_test,ypredTest))
```
