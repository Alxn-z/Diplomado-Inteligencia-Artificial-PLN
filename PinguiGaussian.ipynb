```python
from sklearn import datasets
import pandas as pd
```

```python
pingui=pd.read_csv('/content/penguins_size.csv')
```

```python
pingui
```

```python
pingui=pingui.replace(to_replace=['MALE','FEMALE'],value=[0,1])
```

```python
pingui.dropna(inplace=True)
```

```python
pingui.info()
```

```python
pingui.sample(200)
```

```python
pingui.groupby("sex").count()
```

```python
pingui['sex']=pingui['sex'].replace(to_replace=['MALE','FEMALE'],value=[0,1])
pingui['sex']=pingui['sex'].replace(to_replace=['.'],value=[0])
```

```python
pingui['island']=pingui['island'].replace(to_replace=['Torgersen','Dream','Biscoe'],value=[0,1,2])
```

```python
pingui
```

```python
pingui.describe()
```

```python
X=pingui.drop(columns=['species'],axis=1)
y=pingui['species']
```

```python
X.info()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc=StandardScaler()
```

```python
sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.4,stratify=y,random_state=42)
```

```python
x_test.shape
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
cnBG=GaussianNB()
```

```python
#se aplica la distribución "normal"
cnBG.fit(x_train,y_train)
```

```python
ypredTest=cnBG.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
mx=confusion_matrix(y_test, ypredTest)
mx
```

```python
import seaborn as sns
```

```python
sns.heatmap(mx,annot=True)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, ypredTest))
```
