```python
import pandas as pd
```

```python
plantas=pd.read_csv('/content/plant_growth_data.csv')
plantas
```

```python
cambiarSuelo=({'clay':0,'sandy':1,'loam':2})
plantas['Soil_Type']=plantas['Soil_Type'].map(cambiarSuelo)
```

```python
cambiarAgua=({'daily':0,'bi-weekly':1,'weekly':2})
plantas['Water_Frequency']=plantas['Water_Frequency'].map(cambiarAgua)
```

```python
cambiarF=({'none':0,'chemical':1,'organic':2})
plantas['Fertilizer_Type']=plantas['Fertilizer_Type'].map(cambiarF)
```

```python
plantas
```

```python
X=plantas.drop('Growth_Milestone',axis=1)
y=plantas['Growth_Milestone']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
!pip install catboost
```

```python
from catboost import CatBoostClassifier
```

```python
modelo = CatBoostClassifier(
iterations=100,
learning_rate=0.1,
depth=3,
random_seed=42,
verbose=0 # no mostrar la salida durante el fit
)
```

```python
modelo.fit(x_train, y_train)
```

```python
ypredtest=modelo.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
mx=confusion_matrix(y_test,ypredtest)
mx
```

```python
print(classification_report(y_test,ypredtest))
```
