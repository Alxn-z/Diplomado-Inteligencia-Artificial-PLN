```python
import pandas as pd
```

```python
autos=pd.read_csv('/content/FuelConsumptionCo2.csv')
```

```python
autos.sample(20)
```

```python
X = autos[["ENGINESIZE", "CYLINDERS", "FUELCONSUMPTION_CITY",
            "FUELCONSUMPTION_HWY", "FUELCONSUMPTION_COMB", "FUELCONSUMPTION_COMB_MPG"]]
y = autos[["CO2EMISSIONS"]]
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
#media cero desvi.. 1
scaler = StandardScaler()
```

```python
XT= scaler.fit_transform(X)
```

```python
XT
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(XT, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import Ridge
```

```python
ridge=Ridge(alpha=0.5)
```

```python
#calculé cuáles de las variables aportan más al modelo
ridge.fit(X_train,y_train)
```

```python
ridge.coef_
```

```python
#predijo
y_pred=ridge.predict(X_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
import numpy as np
```

```python
print(r2_score(y_test,y_pred))
```

```python
mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(mean_squared_error(y_test,y_pred))
```
