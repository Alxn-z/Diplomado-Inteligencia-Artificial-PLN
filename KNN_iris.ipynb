```python
# Librerías necesarias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import label_binarize
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import roc_curve, auc
```

```python
# 1. Cargar el dataset Iris
iris = load_iris()
X = iris.data       # Características (4 columnas: largo/alto sépalo y pétalo)
y = iris.target     # Etiquetas (clases 0, 1, 2)
```

```python
# 2. Binarizar las clases para poder graficar curva ROC multiclase
# Convierte: 0 → [1,0,0], 1 → [0,1,0], 2 → [0,0,1]
y_bin = label_binarize(y, classes=[0, 1, 2])
n_classes = y_bin.shape[1]  # En Iris hay 3 clases
```

```python
# 3. Dividir el dataset en entrenamiento (70%) y prueba (30%)
X_train, X_test, y_train, y_test = train_test_split(X, y_bin, test_size=0.3, random_state=42)

```

```python
# 4. Crear y entrenar el modelo multiclase con estrategia One-vs-Rest
# LogisticRegression por sí solo no acepta etiquetas binarizadas,
# así que usamos OneVsRestClassifier para adaptar el entrenamiento
clf = OneVsRestClassifier(LogisticRegression(max_iter=1000))
clf.fit(X_train, y_train)
```

```python
# 5. Obtener las probabilidades predichas para cada clase
# y_score tendrá una columna por clase con su probabilidad
y_score = clf.predict_proba(X_test)
```

```python
# Calcula FPR, TPR y AUC para cada clase
# fpr = tasa de falsos positivos
# tpr = tasa de verdaderos positivos
# auc = área bajo la curva ROC
fpr = dict()
tpr = dict()
roc_auc = dict()

for i in range(n_classes):
    # Compara etiquetas verdaderas con probabilidades predichas para cada clase
    fpr[i], tpr[i], _ = roc_curve(y_test[:, i], y_score[:, i])
    roc_auc[i] = auc(fpr[i], tpr[i])  # Calcula el área bajo la curva ROC
```

```python
# Grafica todas las curvas ROC
plt.figure(figsize=(8, 6))
colors = ['blue', 'green', 'red']

for i in range(n_classes):
    plt.plot(fpr[i], tpr[i], color=colors[i],
             label=f'Clase {i} (AUC = {roc_auc[i]:.2f})')

# Línea diagonal de referencia (azar)
plt.plot([0, 1], [0, 1], 'k--', lw=1, label='Azar (AUC = 0.5)')

# Configuración del gráfico
plt.xlabel('Tasa de Falsos Positivos (FPR)')
plt.ylabel('Tasa de Verdaderos Positivos (TPR)')
plt.title('Curvas ROC por Clase - Dataset Iris')
plt.legend(loc='lower right')
plt.grid(True)
plt.show()
```
